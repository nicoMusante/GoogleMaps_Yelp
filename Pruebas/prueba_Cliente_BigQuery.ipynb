```python
from google.cloud import bigquery
from google.oauth2 import service_account 
import pandas as pd
import json
```

```python
# Asumiendo que 'df' es tu DataFrame y 'columna_dict' es la columna de tipo dict
df['resp'] = df['resp'].apply(json.dumps)
```

```python
# Asumiendo que 'df' es tu DataFrame y 'columna_dict' es la columna de tipo dict
df['pics'] = df['pics'].apply(json.dumps)
```

```python
# Nombre de la columna que quieres verificar
columna_a_verificar = 'pics'

# Recorrer la columna y verificar si hay datos de tipo lista
for index, valor in enumerate(df[columna_a_verificar]):
    if isinstance(valor, list):
        print(f"Encontrado dato de tipo lista en la fila {index}, valor: {valor}")

# Resultado:
# Encontrado dato de tipo lista en la fila 2, valor: [3, 4]
```

```python
# Proporciona la ruta al archivo de clave de tu cuenta de servicio
credentials = service_account.Credentials.from_service_account_file(
    '../credenciales.json')

# Crea un cliente de BigQuery con tus credenciales
client = bigquery.Client(credentials=credentials, project=credentials.project_id)

# Reemplaza con tu ID de conjunto de datos y tabla
table_id = 'proyecto-final-henry-412703.Datawarehouse_PF.reviews-alabama'

# Carga el DataFrame en BigQuery
job = client.load_table_from_dataframe(df, table_id)

# Espera a que la carga se complete
job.result()

print("Carga completada.")
```
